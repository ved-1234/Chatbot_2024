```python
#Extracting UG programs data, table 1
import nest_asyncio
from requests_html import AsyncHTMLSession
from bs4 import BeautifulSoup

# Apply the patch for nested asyncio loops
nest_asyncio.apply()

async def fetch_table_data_1(url):
    session = AsyncHTMLSession()
    response = await session.get(url)
    await response.html.arender()  # Render JavaScript

    soup = BeautifulSoup(response.html.html, 'html.parser')

    # Find the first table on the page
    tables_1 = soup.find_all('table')
    table_1 = tables_1[0]

    # Initialize a list to store the extracted data
    table_data_1 = []

    table_rows_1 = table_1.find_all('tr')
    for row_1 in table_rows_1:
        cols_1 = row_1.find_all('td')
        if len(cols_1) == 2:  # Ensure we have exactly two columns
            name_of_the_course_1 = cols_1[0].text.strip().replace('Name of the course', '').strip()
            intake_1 = cols_1[1].text.strip().replace('Approved no. of intake', '').strip()

            # Append the extracted data to the table_data list
            table_data_1.append({
                'Name of the course': name_of_the_course_1,
                'Intake': intake_1
            })

    return table_data_1

async def main():
    url = 'https://vcet.edu.in/courses-and-intake/'
    table_data_1 = await fetch_table_data_1(url)
    return table_data_1

# Run the async function
import asyncio
result_1 = asyncio.run(main())

# Print the extracted data for Table 1
print("Under Graduate Program ( UG):")
for entry_1 in result_1:
    print(entry_1)

#Extracting Post Graduate Program ( PG) programs data, table 2
async def fetch_table_data_2(url):
    session = AsyncHTMLSession()
    response = await session.get(url)
    await response.html.arender()  # Render JavaScript

    soup = BeautifulSoup(response.html.html, 'html.parser')

    # Find the first table on the page
    tables_2 = soup.find_all('table')
    table_2 = tables_2[1]

    # Initialize a list to store the extracted data
    table_data_2 = []

    table_rows_2 = table_2.find_all('tr')
    for row_2 in table_rows_2:
        cols_2 = row_2.find_all('td')
        if len(cols_2) == 2:  # Ensure we have exactly two columns
            name_of_the_course_2 = cols_2[0].text.strip().replace('Name of the course', '').strip()
            intake_2 = cols_2[1].text.strip().replace('Approved no. of intake', '').strip()

            # Append the extracted data to the table_data list
            table_data_2.append({
                'Name of the course': name_of_the_course_2,
                'Intake': intake_2
            })

    return table_data_2

async def main():
    url = 'https://vcet.edu.in/courses-and-intake/'
    table_data_2 = await fetch_table_data_2(url)
    return table_data_2

# Run the async function
import asyncio
result_2 = asyncio.run(main())

# Print the extracted data for Table 1
print("Post Graduate Program ( PG):")
for entry_2 in result_2:
    print(entry_2)

#Extracting Management Course Program data, table 3
async def fetch_table_data_3(url):
    session = AsyncHTMLSession()
    response = await session.get(url)
    await response.html.arender()  # Render JavaScript

    soup = BeautifulSoup(response.html.html, 'html.parser')

    # Find the first table on the page
    tables_3 = soup.find_all('table')
    table_3 = tables_3[2]

    # Initialize a list to store the extracted data
    table_data_3 = []

    table_rows_3 = table_3.find_all('tr')
    for row_3 in table_rows_3:
        cols_3 = row_3.find_all('td')
        if len(cols_3) == 2:  # Ensure we have exactly two columns
            name_of_the_course_3 = cols_3[0].text.strip().replace('Name of the course', '').strip()
            intake_3 = cols_3[1].text.strip().replace('Approved no. of intake', '').strip()

            # Append the extracted data to the table_data list
            table_data_3.append({
                'Name of the course': name_of_the_course_3,
                'Intake': intake_3
            })

    return table_data_3

async def main():
    url = 'https://vcet.edu.in/courses-and-intake/'
    table_data_3 = await fetch_table_data_3(url)
    return table_data_3

# Run the async function
import asyncio
result_3 = asyncio.run(main())

# Print the extracted data for Table 1
print("Management Course Program:")
for entry_3 in result_3:
    print(entry_3)
```

```python
# Load documents

file_path = '/content/combinedData.txt'  # Replace with the actual path to your text file

# Reading the entire content of the file into a variable
with open(file_path, 'r', encoding='utf-8') as file:
    data = file.read()

# Print the data to verify
print(data)

# Now `data` variable contains the entire content of the text file
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=7000, chunk_overlap=200)
docs = text_splitter.create_documents([data])
```

```python
# Load the Hugging Face model for embeddings
from langchain.embeddings import HuggingFaceEmbeddings
from sentence_transformers import SentenceTransformer
```

```python
# Initialize embeddings using the updated HuggingFaceEmbeddings class
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
import pickle
import faiss
from langchain.vectorstores import FAISS
```

```python
# Create FAISS vector store from documents
vectorStore = FAISS.from_documents(docs, embeddings)
```

```python
# Save the vector store
with open("faiss_store.pkl", "wb") as f:
    pickle.dump(vectorStore, f)
```

```python
# Convert the vector store to a retriever
retriever = vectorStore.as_retriever()
```

```python
from transformers import pipeline
# Initialize a local language model using Hugging Face's pipeline for QA
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")
```

```python
# Define a function to use the QA pipeline with the retriever
def answer_query(query):
    # Retrieve relevant documents
    relevant_docs = retriever.get_relevant_documents(query)
    # Combine the texts of the retrieved documents
    context = " ".join(doc.page_content for doc in relevant_docs)
    # Use the QA pipeline to answer the query based on the context
    result = qa_pipeline(question=query, context=context)
    return result['answer']
```

```python
# Step 9: Test the setup with a sample query
query = "What is the Email ID of college?"
result = answer_query(query)

# Print the result
print(result)
```
